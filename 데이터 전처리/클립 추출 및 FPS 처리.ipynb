구매 fps 3으로 변경

```python
import os
import shutil
from moviepy.editor import VideoFileClip

# 원본 파일이 있는 디렉토리 경로
source_directory = "/content/drive/MyDrive/DL/output_all/data_1"
# 변환된 파일을 저장할 디렉토리 경로
target_directory = "/content/drive/MyDrive/DL/output_all/data_2"

# 목표 FPS 설정
target_fps = 3

# 디렉토리 내 모든 파일 이름을 반복 처리
for filename in os.listdir(source_directory):
    if filename.endswith((".mp4", ".avi", ".mov")):  # 비디오 파일 확장자 필터링
        original_file_path = os.path.join(source_directory, filename)

        # SMA_purchase로 시작하는 파일의 FPS 변경
        if filename.startswith("SMA_purchase"):
            # VideoFileClip을 사용하여 비디오 파일 열기
            clip = VideoFileClip(original_file_path)

            # 비디오의 FPS 정보 가져오기
            original_fps = clip.fps
            print(f"Original FPS of {filename}: {original_fps}")

            # FPS를 맞춰서 비디오 변환
            if original_fps != target_fps:
                new_file_path = os.path.join(target_directory, filename)

                # 비디오의 모든 프레임을 target_fps로 변환하여 저장
                clip.set_fps(target_fps).write_videofile(new_file_path, codec='libx264', audio_codec='aac')
                print(f"Converted {filename} to {target_fps} FPS and saved to {new_file_path}")
            else:
                # FPS 변환이 필요 없으면 파일을 그대로 복사
                new_file_path = os.path.join(target_directory, filename)
                shutil.copy(original_file_path, new_file_path)
                print(f"Copied {filename} to {new_file_path} without conversion")

            # 작업 후 클립 닫기
            clip.close()
        else:
            # 나머지 파일들은 그대로 복사
            new_file_path = os.path.join(target_directory, filename)
            shutil.copy(original_file_path, new_file_path)
            print(f"Copied {filename} to {new_file_path} without conversion")
```
