```python
import os
import random
import shutil

# 디렉토리 설정
video_dir = '/content/drive/MyDrive/DL/output_all/mediapipe/processed_videos'
csv_dir = '/content/drive/MyDrive/DL/output_all/mediapipe/processed_csvs'
test_video_dir = '/content/drive/MyDrive/DL/output_all/mediapipe/test_videos'
test_csv_dir = '/content/drive/MyDrive/DL/output_all/mediapipe/test_csvs'

# 클래스별 파일 목록 가져오기
broken_files = []
purchase_files = []
theft_files = []

for filename in os.listdir(csv_dir):
    if filename.endswith(".csv"):
        if "broken" in filename:
            broken_files.append(filename)
        elif "purchase" in filename:
            purchase_files.append(filename)
        elif "theft" in filename:
            theft_files.append(filename)

# 샘플링 비율 설정
sample_ratio = 0.2

# 클래스별 샘플링
broken_sample = random.sample(broken_files, int(len(broken_files) * sample_ratio))
purchase_sample = random.sample(purchase_files, int(len(purchase_files) * sample_ratio))
theft_sample = random.sample(theft_files, int(len(theft_files) * sample_ratio))

# 테스트 디렉토리가 없으면 생성
os.makedirs(test_video_dir, exist_ok=True)
os.makedirs(test_csv_dir, exist_ok=True)

# 샘플 파일을 테스트 디렉토리로 이동
def move_files(sample_list, source_csv_dir, source_video_dir, dest_csv_dir, dest_video_dir):
    for csv_file in sample_list:
        base_filename = csv_file.replace(".mp4.csv", "")
        video_file = f"processed_{base_filename}.mp4"
        # csv 파일 이동
        shutil.move(os.path.join(source_csv_dir, csv_file), os.path.join(dest_csv_dir, csv_file))
        # 비디오 파일 이동
        if os.path.exists(os.path.join(source_video_dir, video_file)):
            shutil.move(os.path.join(source_video_dir, video_file), os.path.join(dest_video_dir, video_file))
        else:
            print(f"Video file {video_file} not found for CSV file {csv_file}")

move_files(broken_sample, csv_dir, video_dir, test_csv_dir, test_video_dir)
move_files(purchase_sample, csv_dir, video_dir, test_csv_dir, test_video_dir)
move_files(theft_sample, csv_dir, video_dir, test_csv_dir, test_video_dir)

print("샘플링 완료. 파일이 테스트 디렉토리로 이동되었습니다.")
```
